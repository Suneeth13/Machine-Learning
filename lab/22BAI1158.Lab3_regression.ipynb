```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
y = data['GPA']
x1 = data['SAT']
```

```python
plt.scatter(x1,y)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 0.0017*x1 + 0.275
x2 = np.array(x1)
y2 = np.array(yhat)
fig = plt.plot(x2,y2,lw = 4, c = 'orange', label = 'regression line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/parenthood.csv")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

ph_df = pd.DataFrame({
    'dadsleep': data['dan.sleep'],
    'dadgrump': data['dan.grump']
})
@widgets.interact(slope=widgets.IntSlider(min=-20, max=20, step=1, value=0), intercept=widgets.IntSlider(min=0, max=200, step=1, value=50))
def plot_grumpiness(slope, intercept):
    predict = intercept+slope*ph_df.dadsleep
    resid = ph_df.dadgrump-predict
    fig, ax = plt.subplots(figsize=(9,6))

    x1 = np.linspace(ph_df.dadsleep.min(), ph_df.dadsleep.max(),400)
    y1 = intercept+slope*x1

    ax.plot(ph_df.dadsleep, ph_df.dadgrump,'ko',markersize=4)
    ax.plot(ph_df.dadsleep, ph_df.dadgrump-resid,'o',markersize=4,markeredgecolor='r', markeredgewidth=.4, markerfacecolor='white')

    ax.plot(x1,y1,'-',color='steelblue',linewidth=1)
    plt.xlabel("dad sleep")
    plt.ylabel("dad grumpiness")

    ax.vlines(ph_df.dadsleep, ph_df.dadgrump, ph_df.dadgrump-resid,'r',linewidth=0.5)
    plt.show()
```

```python
# Question 3

# Observations from the Plot:

# The best fit line is the one that minimizes the overall amount of red in the plot, i.e., the sum of squared residuals.
# A good fit is observed when the red points (residuals) are evenly distributed around the blue line (the regression line).
# The best fit line captures the trend in the data and provides a good representation of the relationship between dad sleep and dad grumpiness.
```

```python
#Question 4
from sklearn.linear_model import LinearRegression
import seaborn as sns

# Fit the linear regression model
reg_model = LinearRegression()
reg_model.fit(ph_df[['dadsleep']], ph_df['dadgrump'])

# Plot the regression line
sns.regplot(x='dadsleep', y='dadgrump', data=ph_df, scatter_kws={'s': 20})
plt.xlabel("dad sleep")
plt.ylabel("dad grumpiness")
plt.title("Linear Regression Fit")

plt.show()
```

```python
#Question 5:
# Comparison of Output Parameters:

# The output parameters from the code (reg_model.coef_ and reg_model.intercept_) should be compared with the slope and intercept you obtained manually.
# reg_model.coef_ corresponds to the coefficient of dadsleep, and reg_model.intercept_ corresponds to the y-intercept.
#Question 6:
plt.scatter(ph_df['dadsleep'], ph_df['dadgrump'], color='black')
plt.plot(ph_df['dadsleep'], reg_model.predict(ph_df[['dadsleep']]), color='blue', linewidth=2)
plt.xlabel("dad sleep")
plt.ylabel("dad grumpiness")
plt.title("Linear Regression Model with Data")

plt.show()
```

```python
# Question 7
# Interpretation of Dadsleep Coefficient:

# The coefficient of dadsleep represents the change in dad grumpiness for a one-unit increase in dadsleep.
# In this case, if the dadsleep coefficient is positive, it indicates that as the hours of dad sleep increase, dad grumpiness also tends to increase.
```

```python
#Question 8
hours_of_sleep = 6
predicted_grumpiness = reg_model.predict([[hours_of_sleep]])
print(f"Predicted grumpiness for {hours_of_sleep} hours of sleep: {predicted_grumpiness[0]}")
```
